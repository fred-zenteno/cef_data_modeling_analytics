```python
final_enrollment_el_df = pd.concat([enrollment_el_2023_df, enrollment_el_2022_df, enrollment_el_2021_df, enrollment_el_2020_df, enrollment_el_2019_df, enrollment_el_2018_df])
final_enrollment_el_df
```

```python
final_enrollment_el_df.loc[final_enrollment_el_df['AggLevel'] == 'T', 'CountyCode'] = '00'
final_enrollment_el_df.loc[final_enrollment_el_df['AggLevel'] == 'T', 'DistrictCode'] = '00000'
final_enrollment_el_df.loc[final_enrollment_el_df['AggLevel'] == 'T', 'SchoolCode'] = '0000000'

final_enrollment_el_df['Charter'] = final_enrollment_el_df['Charter'].fillna('NV').replace('', 'NV')
```

```python
final_enrollment_el_df.loc[final_enrollment_el_df['AggLevel'] == 'D', 'SchoolCode'] = '0000000'
```

```python
final_enrollment_el_df[final_enrollment_el_df['AggLevel'] == 'D']

# D = Local educational agency totals (includes districts and direct funded charter schools)
# O = County totals
# S = School totals
# T = State totals
```

```python
final_enrollment_el_df.Charter.unique()
```

```python
# Columns to convert to integer
convert_columns = [
    'EO', 'IFEP', 'EL', 'RFEP', 'TBD', 'TotalEnrollment', 'AR', 'LTEL',
    'EL4+', 'EL03Y', 'EL45Y', 'EL6+Y', 'Total-EE'
]

# Convert specified columns from strings to integers (handling missing or malformed data)
for column in convert_columns:
    final_enrollment_el_df[column] = pd.to_numeric(final_enrollment_el_df[column], errors='coerce').fillna(0).astype(int)

# Group by columns
group_columns = [
    'AcademicYear', 'CountyCode', 'DistrictCode', 'SchoolCode',  'AggLevel', 'Gender'
]

# Group the data and sum the specified columns
final_enrollment_el_df = final_enrollment_el_df.groupby(group_columns)[convert_columns].sum().reset_index()
final_enrollment_el_df
```

```python
# Step 1: Create a set of unique school codes from dataframe2 for efficient lookup
wcc_schools_list = set(wcc_schools_df['School Code'])

# Step 2: Use the .isin() method to check if each school code in dataframe1 is in the set of schools from dataframe2
final_enrollment_el_df['is_wcc_entity'] = final_enrollment_el_df['SchoolCode'].isin(wcc_schools_list)
```

```python
# Existing conditions for "Aggregate Level" and "is_wcc_entity"
condition1 = final_enrollment_el_df['AggLevel'].isin(['T', 'D'])
condition2 = final_enrollment_el_df['is_wcc_entity']

# Combine all conditions using "OR" (|) for the initial conditions
combined_condition = (condition1 | condition2) 

# Filter dataframe1 using the combined condition
final_filtered_enrollment_el_df = final_enrollment_el_df[combined_condition]
```

```python
final_filtered_enrollment_el_df
```

```python
# Define a dictionary with the mappings from old values to new values
replacement_dict = {
    'ALL': 'TA',
    'F': 'GF',
    'M': 'GM',
    'X': 'GX'
}

# Replace the values in the 'Gender' column
final_filtered_enrollment_el_df['Gender'] = final_filtered_enrollment_el_df['Gender'].replace(replacement_dict)
final_filtered_enrollment_el_df
```

```python
# Dictionary mapping old column names to new ones
rename_dict = {
    'AcademicYear': 'academic_year',
    'CountyCode': 'county_code',
    'DistrictCode': 'district_code',
    'SchoolCode': 'school_code',
    'AggLevel': 'aggregate_level',
    'Gender': 'reporting_category' 
}

# Rename the columns using the dictionary
final_filtered_enrollment_el_df = final_filtered_enrollment_el_df.rename(columns=rename_dict)
final_filtered_enrollment_el_df
```

```python
# Columns you want to remove
columns_to_drop = ['is_wcc_entity' ]

# Drop the specified columns
final_filtered_enrollment_el_df = final_filtered_enrollment_el_df.drop(columns=columns_to_drop)
final_filtered_enrollment_el_df
```

```python
final_filtered_enrollment_el_df.to_csv('final_long_datasets_domain/final_enrollment_el_df.csv', index=False)
```
