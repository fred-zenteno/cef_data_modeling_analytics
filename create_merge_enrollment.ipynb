```python
import pandas as pd
```

```python
cum_enrollment_df = pd.read_csv('final_long_datasets_domain/final_cum_enrollment_df.csv', encoding='ISO-8859-1',dtype=str)
pd.set_option('display.max_columns', None)
cum_enrollment_df
```

```python
el_enrollment_df = pd.read_csv('final_long_datasets_domain/final_enrollment_el_df.csv', encoding='ISO-8859-1',dtype=str)
pd.set_option('display.max_columns', None)
el_enrollment_df
```

```python
complete_enrollment_df = pd.merge(cum_enrollment_df, el_enrollment_df, on=['academic_year', 'county_code', 'district_code', 'school_code', 'aggregate_level', 'reporting_category'], how='left')
complete_enrollment_df
```

```python
# Define the mapping from metric_id
year_mapping = {
    '2017-18': '2018',
    '2018-19': '2019',
    '2019-20': '2020',
    '2020-21': '2021',
    '2021-22': '2022',
    '2022-23': '2023'
}

# Use .loc to ensure you're modifying the original DataFrame
complete_enrollment_df.loc[:, 'academic_year'] = complete_enrollment_df['academic_year'].astype(str).replace(year_mapping)
```

```python
complete_enrollment_df['aggregate_level'] = complete_enrollment_df['aggregate_level'].replace({'T': 'State', 'D': 'District', 'S': 'School'})
```

```python
complete_enrollment_df
```

```python
complete_enrollment_df.to_csv('tableau_ready_files/enrollment_fact.csv', index=False)
```
