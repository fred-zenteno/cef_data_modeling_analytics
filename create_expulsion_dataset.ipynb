```python
final_expulsion_df = pd.concat([expulsion23_df, expulsion22_df, expulsion21_df, expulsion20_df, expulsion19_df, expulsion18_df])
final_expulsion_df
```

```python
# Step 1: Create a set of unique school codes from the wcc school list for efficient lookup
wcc_schools_list = set(wcc_schools_df['School Code'])

# Step 2: Use the .isin() method to check if each school code in dataframe1 is in the set of schools from dataframe2
final_expulsion_df['is_wcc_entity'] = final_expulsion_df['school_code'].isin(wcc_schools_list)

final_expulsion_df
```

```python
# Existing conditions for "Aggregate Level" and "is_wcc_entity"
condition1 = final_expulsion_df['aggregate_level'].isin(['T', 'D'])
condition2 = final_expulsion_df['is_wcc_entity']

# New conditions to exclude rows with "*" in the chronic absenteeism columns
condition3 = ~final_expulsion_df['cumulative_enrollment'].str.contains('*', regex=False)

# Combine all conditions using "AND" (&) for the chronic absenteeism conditions and "OR" (|) for the initial conditions
combined_condition = (condition1 | condition2) & (condition3)

# Filter dataframe1 using the combined condition
final_filtered_expulsion_df = final_expulsion_df[combined_condition]

final_filtered_expulsion_df
```

```python
final_filtered_expulsion_df.to_csv('final_long_datasets_domain/final_expulsion_df.csv', index=False)
```
