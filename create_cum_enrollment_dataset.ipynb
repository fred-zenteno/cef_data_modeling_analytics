```python
final_enrollment_df = pd.concat([enrollment_2023_df, enrollment_2022_df, enrollment_2021_df, enrollment_2020_df, enrollment_2019_df, enrollment_2018_df])
final_enrollment_df
```

```python
# Step 1: Create a set of unique school codes from wcc schools for efficient lookup
wcc_schools_list = set(wcc_schools_df['School Code'])

# Step 2: Use the .isin() method to check if each school code in dataframe1 is in the set of schools 
final_enrollment_df['is_wcc_entity'] = final_enrollment_df['SchoolCode'].isin(wcc_schools_list)
```

```python
# Existing conditions for "Aggregate Level" and "is_wcc_entity"
condition1 = final_enrollment_df['AggregateLevel'].isin(['T', 'D'])
condition2 = final_enrollment_df['is_wcc_entity']

# New conditions to exclude rows with "*" in the chronic absenteeism columns
condition3 = ~final_enrollment_df['CumulativeEnrollment'].str.contains('*', regex=False)

# Combine all conditions using "AND" (&) for the chronic absenteeism conditions and "OR" (|) for the initial conditions
combined_condition = (condition1 | condition2) & (condition3)

# Filter dataframe1 using the combined condition
final_filtered_enrollment_df = final_enrollment_df[combined_condition]
```

```python
final_filtered_enrollment_df.columns = [
    'academic_year',
    'aggregate_level',
    'county_code',
    'district_code',
    'school_code',
    'county_name',
    'district_name',
    'school_name',
    'charter_yn',
    'reporting_category',
    'cumulative_enrollment', 
    'is_wcc_entity'
]

final_filtered_enrollment_df
```

```python
# Condition for dropping rows
# Keep rows where 'aggregate_level' is NOT 'D' OR 'charter_yn' IS 'All'
condition = ~((final_filtered_enrollment_df['aggregate_level'] == 'D') & (final_filtered_enrollment_df['charter_yn'] != 'All'))

# Apply the condition to the DataFrame
final_filtered_enrollment_df = final_filtered_enrollment_df[condition]
```

```python
# Columns you want to remove
columns_to_drop = ['county_name', 'district_name', 'school_name', 'charter_yn', 'is_wcc_entity' ]

# Drop the specified columns
final_filtered_enrollment_df = final_filtered_enrollment_df.drop(columns=columns_to_drop)
```
