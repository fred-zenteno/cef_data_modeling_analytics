```python
import pandas as pd
```

# 2023

```python
chronicabsenteeism23_df = pd.read_csv('chronic_absenteeism_csv_files/chronicabsenteeism23.csv', encoding='ISO-8859-1',dtype=str)
chronicabsenteeism23_df
```

```python
# Step 1: Create a set of unique school codes from dataframe2 for efficient lookup
wcc_schools_list = set(wcc_schools_df['School Code'])

# Step 2: Use the .isin() method to check if each school code in dataframe1 is in the set of schools from dataframe2
chronicabsenteeism23_df['is_wcc_entity'] = chronicabsenteeism23_df['School Code'].isin(wcc_schools_list)

chronicabsenteeism23_df
```

```python
# Existing conditions for "Aggregate Level" and "is_wcc_entity"
condition1 = chronicabsenteeism23_df['Aggregate Level'].isin(['T', 'D'])
condition2 = chronicabsenteeism23_df['is_wcc_entity']

# New conditions to exclude rows with "*" in the chronic absenteeism columns
condition3 = ~chronicabsenteeism23_df['ChronicAbsenteeismEligibleCumulativeEnrollment'].str.contains('*', regex=False)
condition4 = ~chronicabsenteeism23_df['ChronicAbsenteeismCount'].str.contains('*', regex=False)
condition5 = ~chronicabsenteeism23_df['ChronicAbsenteeismRate'].str.contains('*', regex=False)

# Combine all conditions using "AND" (&) for the chronic absenteeism conditions and "OR" (|) for the initial conditions
combined_condition = (condition1 | condition2) & (condition3 & condition4 & condition5)

# Filter dataframe1 using the combined condition
final_filtered_dataframe23 = chronicabsenteeism23_df[combined_condition]

final_filtered_dataframe23
```

```python
# Conditions for rows to be removed
condition1 = (final_filtered_dataframe23['Aggregate Level'] == 'T') & (final_filtered_dataframe23['DASS'] != 'All')
condition2 = (final_filtered_dataframe23['Aggregate Level'] == 'D') & (final_filtered_dataframe23['DASS'] != 'All')

# Combine conditions with an OR '|', since rows meeting either condition should be removed
conditions_combined = condition1 | condition2

# Invert the conditions to keep rows that do NOT meet either condition
final_filtered_dataframe23 = final_filtered_dataframe23[~conditions_combined]
final_filtered_dataframe23
```

```python
grouped_counts = final_filtered_dataframe23.groupby(['Aggregate Level', 'DASS']).size().reset_index(name='count')
grouped_counts
```
