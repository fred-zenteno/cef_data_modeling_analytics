```python
caaspp_df = pd.read_csv('final_long_datasets_domain/final_caaspp_all_grades_df.csv', encoding='ISO-8859-1',dtype=str)

caaspp_df
```

```python
# Function to determine the aggregate level
def determine_aggregate_level(row):
    if row['county_code'] == '00':
        return 'T'
    elif row['district_code'] != '00000' and row['school_code'] == '0000000':
        return 'D'
    elif row['district_code'] != '00000' and row['school_code'] != '0000000':
        return 'S'
    else:
        return 'X'

# Apply the function to each row to create the new column
caaspp_df['aggregate_level'] = caaspp_df.apply(determine_aggregate_level, axis=1)
```

```python
# Assuming caaspp_df is your DataFrame and it already contains the 'type_id' column

# Function to determine charter_yn based on type_id
def get_charter_yn(type_id):
    if pd.isnull(type_id) or type_id == '':
        return 'NV'
    mapping = {
        '4': 'All',
        '6': 'All',
        '7': 'No',
        '9': 'Yes',
        '10': 'Yes'
    }
    return mapping.get(type_id, 'Other')  # 'Other' is a default value if type_id is not in the mapping

# Apply the function to each row in the 'type_id' column to create the new 'charter_yn' column
caaspp_df['charter_yn'] = caaspp_df['type_id'].apply(get_charter_yn)
```

```python
# List of columns to clean
columns_to_clean = [
    'area_1_pct_above_std', 'area_1_pct_near_std', 'area_1_pct_below_std',
    'area_2_pct_above_std', 'area_2_pct_near_std', 'area_2_pct_below_std',
    'area_3_pct_above_std', 'area_3_pct_near_std', 'area_3_pct_below_std',
    'area_4_pct_above_std', 'area_4_pct_near_std', 'area_4_pct_below_std'
]

# Replace any value containing '*' with an empty string (or another placeholder if needed)
for column in columns_to_clean:
    caaspp_df[column] = caaspp_df[column].replace(to_replace=r'.*\*.*', value='', regex=True)
    
```
